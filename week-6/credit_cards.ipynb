```python
# rename the "Card Type Full Name" and Card Holder's Name columns
# to "Type" and "Client"

data.rename(columns={"Card Type Full Name":"Type", "Card Holder's Name":"Client"}, inplace=True)
data.head(5)
```

```python
# Show card type, bank, client for all cases where credit limit > 70000

data[(data["Credit Limit"] > 70000)]
```

```python
# find all clients, billing date, card number for clients of Chase bank

data[(data["Issuing Bank"] == "Chase")]
```

```python
# Show all clients card numbers, credit limit for visa and with credit limit > 100000

data[(data["Type"] == "Visa") & (data["Credit Limit"] > 100000)]

# this needs to be fixed
```

```python
# get all clients, issuing bank, credit limit of all those that are Master Card and expiry date is after 2020

# this needs to be fixed

after_2020 = data[(data["Type"] == "Master Card") & (data["Expiry Date"].apply(lambda x: x[3:]) >= "2021")]

after_2020[["Client", "Issuing Bank", "Credit Limit"]]
```
