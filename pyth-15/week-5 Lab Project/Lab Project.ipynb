```python
import os
import csv

first_start = True


def add_new_entry():
    title = ""
    expense_type = ""
    amount = 0
    date = ""
    income_or_expense = 1

    # if not os.path.exists("./budget_output.csv"):
    with open("./budget_output.csv", "a") as file:

        writer = csv.writer(file)

        title = input("Please enter a title: ")

        while expense_type != "e" and expense_type != "i":
            expense_type = input("Is this an [E]xpense or [I]ncome? ").lower()

        expense_type = "Income" if expense_type == "i" else "Expense"
        income_or_expense = 1 if expense_type == "Income" else -1

        amount = int(input("Please enter an amount: ")) * income_or_expense

        date = input("Please enter a date: ")

        writer.writerow([title, expense_type, amount, date])
      #  file.flush()

        

def display_total_account_bal():
    total = 0
    average = 0
    index = 0

    with open("./budget_output.csv", "r") as file:

        rows = []
        reader = csv.reader(file, delimiter=",")

        for line in reader:
            rows.append(line)
        
        for row in rows:
                total += float(row[2])
                index += 1
            
        average = total // index

        print(f"Your total account balance for your account is ${total} and the average transaction size is ${average}")

def display_all_account_entries():
    

    with open("./budget_output.csv", "r") as file:

        rows = []
        reader = csv.reader(file, delimiter=",")

        for line in reader:
            rows.append(line)

        print("All transactions made:\n")

        for row in rows:
            print(f"{row[0]}\t{row[3]}\t{row[1]}:\t{row[2]}")
        
        print("\n")


def start():
    command = ""
    while command != "q":
        command = input(
            "Welcome to the Budget Tracker\n\nPress [a] to add a new entry.\nPress [b] to display total account balance.\nPress [c] to view all previous entries.\nPress [q] to quit.").lower()

        if command == "a":
            add_new_entry()
            
        if command == "b":
            display_total_account_bal()
            
        if command == "c":
            display_all_account_entries()

if first_start:
    first_start = False
    start()
```
