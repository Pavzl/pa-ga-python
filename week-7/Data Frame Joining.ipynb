```python
data = {"make": ["Toyota", "Lexus", "BMW", "Hyndai"],   
"Model": ["Rav 4", "LS20", "XS", "Elantra"],    
"Year": [2018, 2015, 2012, 2222],   
"Color": ["White", "Black", "Gerey", "Blue"] }
```

```python
import pandas as pd

cars = pd.DataFrame(data)
```

```python
cars
```

```python
more_data = {"make": ["Toyota", "Lexus", "BMW", "Hyndai"],       
"Year": [0000, 2222, 1111, 5555], 
"Model": ["Corolla", "LR99", "X6", "whatev"],
"Color": ["White", "Black", "Grey", "Blue"] }

more_cars = pd.DataFrame(more_data)
```

```python
# combine 2 datasets

pd.concat([cars, more_cars])
```

```python
price_data = {"Price": [35000, 50000, 54000, 74000], "Seats": [5,5,5,5]}

prices = pd.DataFrame(price_data)
```

```python
pd.concat([cars, prices], axis=1)
```

```python
movies = pd.read_csv("movies.csv")
actors = pd.read_csv("actors.csv")
```

```python
# to get the actor data for movies, join the movies and actors databases

pd.merge(movies, actors, how="inner", on="movie_id")
```

```python
# left join
# this will show all rows, including the ones without data

pd.merge(movies, actors, how="left", on="movie_id")
```

```python
# right join
# this will show only rows with complete data

pd.merge(movies, actors, how="right", on="movie_id")
```
